```python
READING , WRITING AND APPENDING MODES
```

```python
mode = 'r' is read only
mode = 'w' is write only (will overwrite files or create new)
mode = 'a' is append only (will add on to the files)
mode = 'r+' is reading and writing
mode = 'w+' is writing and reading (will overwrite exsisting files or create new files)

```

```python
%%writefile Sunday2911.txt 
Everyone will attend the mock test - 29/11
Everyone will attend the mock interview - 30/11
If you do not attend the mock - You miss something very important
```

```python
with open ('Sunday2911.txt' , mode = 'r') as f:
    print(f.read())
```

```python
with open ('Sunday2911.txt', mode ='a') as f:
    f.write('\n we will have 15 questions out of which you need to clear atleast 11')
```
