```python
Grade system

Marks →
90–100 = A
80–89 = B
70–79 = C
Below 70 = Fail
```

```python
def marks(x):
   if x in range(90,101):
       print('A')
   elif x in range (80,90):
       print ('B')
   elif x in range (70,80):
       print ('C')    
   else:
        print ("Fail")
```
