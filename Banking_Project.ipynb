```python
class BankAccount:

    def __init__(self, Name, Account_Number, Balance, Pin, Account_type):
        self.Name = Name
        self.Account_Number = Account_Number
        self.balance = Balance         
        self.Pin = str(Pin)
        self.Account_type = Account_type
        self.wrong_attempts = 0        

    def deposit(self, amount):
        if amount > 0:
            self.balance += amount
            print(f"₹{amount} deposited. New balance: ₹{self.balance}")
        else:
            print("Invalid Amount")

    def withdraw(self, amount):
        if amount <= 0:
            print("Invalid Amount")
            return
        
        if amount > self.balance:
            print("Insufficient funds")
        else:
            self.balance -= amount
            print(f"₹{amount} withdrawn. New balance: ₹{self.balance}")

    def check_balance(self):
        print(f"Account Balance: ₹{self.balance}")

    def atm_access(self):
        while self.wrong_attempts < 3:
            pin = input("Enter ATM PIN: ")

            if pin == self.Pin:
                print("PIN Verified")
                self.wrong_attempts = 0
                return True
            else:
                self.wrong_attempts += 1
                print(f"Wrong PIN! Attempts: {self.wrong_attempts}/3")

        print("ATM Blocked due to 3 wrong PIN attempts!")
        return False



class SavingsAccount(BankAccount):

    def __init__(self, name, acc_no, balance, pin):
        super().__init__(name, acc_no, balance, pin, "Savings")
        self.withdraw_limit = 20000

    
    def withdraw(self, amount):
        if amount > self.withdraw_limit:
            print(f"Withdraw limit is ₹{self.withdraw_limit} per transaction!")
            return

        super().withdraw(amount)


class CurrentAccount(BankAccount):

    def __init__(self, name, acc_no, balance, pin):
        super().__init__(name, acc_no, balance, pin, "Current")
```

```python
s = SavingsAccount("Ramu", 1001, 50000, 1234)
c = CurrentAccount("Sachin", 2001, 80000, 4321)

s.deposit(5000)
s.withdraw(15000)
s.check_balance()

print("\nATM Test:")
if s.atm_access():
    s.withdraw(2000)
```
