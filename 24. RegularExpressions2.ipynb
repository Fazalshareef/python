# Character Identifiers

```python
import re
```

```python
text = 'My phone number is 408-555-1234'
text1 = 'Now the phone number has changed but the old one remains for something else 408-555-1234'
```

```python
phone = re.search(r'\d{3}-\d{3}-\d{4}', text1)
```

```python
phone
```

```python
phone.group()
```

# Quantifiers

```python
phone = re.search(r'\d{3}-\d{3}-\d{4}',text)
```

```python
phone
```

```python
phone_number = re.compile(r'(\d{3})-(\d{3})-(\d{4})')
```

```python
phone_number
```

```python
results = re.search(phone_number, text1)
```

```python
results
```

```python
results.group()
```

```python
results.group(0)
```

```python
results.group(1)
```

```python
results.group(2)
```

```python
results.group(4)
```

# Additional Regex Syntax

```python
re.search(r'cat i','The cat is here') 
```

```python
re.search(r'cat|dog','The dog cat is here') # | 'pipeline' > OR operator
```

```python
re.findall(r'at','The cat with the hat sat there on the mat')
```

```python
re.findall(r'.at','The dogcat with the hat at there') #use period as wildcard
```

```python
re.findall(r'...at','The cat in the hat at went splat.')
```

```python
re.findall(r'^\d', '0 The number 1 is the start digit.') # use Caret '^' > everything that starts with a number
```

```python
re.findall(r'^\d', 'Hi, the start digit.')
```

```python
re.findall(r'^\w{2,4}', '5555 Hi is the first digit.')
```

```python
re.findall(r'^\D', ' is 1 digit.')
```

```python
re.findall(r'\w{4}$', '505464 564651+$1%/ ') 
```

```python
re.findall(r'\d$', 'The digit starts with 2') #dollar $ > used to specify the last character is a digit
```

```python
re.findall(r'\d$', '3 The digit starts with 33 and ends with 3000')
```

```python
#exclude the numbers in the pattern
```

```python
phrase = 'there are 3 numbers 3 4 in123side 55556454 in this sentence'
```

```python
pattern = r'[^\d]' # use [^] for exclusions in a group
```

```python
re.findall(pattern,phrase)
```

```python
pattern = r'[^\d]+'
```

```python
re.findall(pattern,phrase)
```

```python
test_phrase = 'This is a string! But it has punctuation. How can we remove it?'
```

```python
clean_line = re.findall(r'[^!.? ]+',test_phrase)
```

```python
clean_line
```

```python
' '.join(clean_line)
```
