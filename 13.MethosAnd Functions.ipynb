```python


LEGB RULES (Nested Statements)

L: Local - Names assigned in any way within a function (def), and not declared global in that function.

E: Enclosing function locals - Names in the local scope od any and all anclosing functions from iner to outer.

G: Global (module) - Names assigned at the top-level of a module file, or declared global in a def within the file.

B: Built-in (Python) - Names preassigned in the built-in names module: open, range, syntaxerror...    
```

```python
name = 'This is an outer string'

def greet():
    name = 'Learning again'
    print('Hi : ' + name)

    def hello():
        name = 'New Learning'
        print ('Hello: ' + name)

        def demo():
            name = 'Learning'
            print('demo : ' + name)

        demo()

    hello()
```
