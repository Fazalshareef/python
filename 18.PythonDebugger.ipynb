```python
Generally, in order to figure out what errors there are in the code, we've used print() to track the error. However, python comes with a
built-in debugger tool that allow to interactively explore variables within mid operation of the python code.
```

```python
x = [1,2,3]
y = 2
z = 3

result = y + z 
print (result)
result2 = x + y
print(y)
print(result2)
```

```python
import pdb
```

```python
x = [1,2,3]
y = 2
z = 3

result = y + z 
print (result)
pdb.set_trace()
result2 = x + y
print(y)
print(result2)
```
