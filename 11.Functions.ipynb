```python
Write a function grade(marks)

90+ → "A"

75–89 → "B"

60–74 → "C"

40–59 → "D"

<40 → "F"

Return the grade.
```

```python
def marks(n):
    if n >= 90:
        return "A"
    elif n >= 75:
        return "B"
    elif n >= 60:
        return "C"
    elif n >= 40:
        return "D"
    else:
        return "F"
```

```python
marks(78)
```

```python
Write a function sum_list(numbers)
It should return the sum of all numbers in the list.
```

```python
def sum_list(numbers):
    total = 0
    for i in numbers:
        total += i
        
    return total
```

```python
print(sum_list(range(1,10)))
```

```python
Write a function count_odd(numbers)
Return how many odd numbers are in the list.
```

```python
def count_odd(numbers):
    count = 0
    for i in numbers:
        if i % 2 != 0:
           count += 1
    
    return count      
```

```python
print(count_odd(range(1,30)))
```
