```python
5Ô∏è‚É£ Leap Year Checker

Take a year (e.g. 2024).
If it‚Äôs divisible by 4 but not 100, or divisible by 400 ‚Üí print "Leap Year", else "Not a Leap Year".
```

```python
year = 2024
if (year % 400 == 0) or (year % 4 == 0 and year % 100 != 100):
    print (year, "Leap year")
else:
    print (year, "not a leap year")
```

```python
Password Strength Checker üîê

Input a password:

If length ‚â• 8 and contains both letters & numbers ‚Üí "Strong Password"

If length ‚â• 6 ‚Üí "Moderate Password"

Else ‚Üí "Weak Password"
```

```python
password = ("pasWord123")
if (len(password) >= 8) and (any(not char.isalnum() for char in password)):
    print ("Strong password")
elif (len(password) >= 6) and (any(not char.isalnum() for char in password)):
    print ("Moderate password")
else:
    print ("Weak password")
```
