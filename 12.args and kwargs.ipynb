```python
def new_function (*args): # args saves input as tuples by default
    for item in args:
        print (item)
```

```python
new_function(50,'a','@','besant',50.5)
```

```python
def my_function (**kwargs): #kwargs saves the input as dictionaries by default
    if 'fruit' in kwargs:
        print ('My fruit of choice is {}' .format(kwargs['fruit']))
    else:
        print ('I did not find any fruit here')
```

```python
my_function(veg = 'tomato', car = 'i20', animal = 'dog', fruit = 'apple')
```

```python
my_function(vegetable='cucumber')
```

```python
def my_new_func(*args, **kwargs):
    print ('I would like to order {} {}'. format(args[-1], kwargs['food']))
```

```python
my_new_func(10,20,30,40,457,45,food='biscuit',fruit='apple',chocolate='kitkat')
```

```python
def my_new_func(**kwargs):
    print(kwargs)
    print('I would like to order {}'.format(kwargs['fruit']))
```

```python
my_new_func(food='biscuit', fruit='apple',chocolate='kitkat')
```
