```python
Errors are bound to happen in your code especially when some else ends up using in an unexcepted way. We can use error handling to attemp 
to plan possible errors.

    In case of any type of error in code, the script will stop.

    We can use Error Handling to let the script continue with other code, even if there is an error.
```

```python
Solution :

We use three key words:

    try: This is the block of code to be attempted (may lead to an error)
    except : The Block of code will execute in case there is any error in try block.
    finally : A finally block of code to be executed, regardless of an error.
```

```python
def add(n1, n2):
    print("previous line")
    print(n1+n2)
    print("next line")
```

```python
add(10, '20')
```

```python
print ('Before try')
try:
    print("line one")
    result = 10 + 10
    print ("next line")
except:
    print('seems like there is an issue with addition')
print('after try')    
```

```python
try:
    result = 10 + 10
    print(result)
except:
    print('Seems like there is an issue with addition')
```

```python
try:
    result = 10 + '10'
    x = 10/0
    print(result)
    print(x)
except:
    print('seems like there is an issue with addition')
else:
    print('Add went well')
    print(result)
    print('Sucessfull')
finally:
    print('Bye')
```
