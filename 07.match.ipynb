```python
students = [("Fazal", 88), ("Ravi", 65), ("John", 45), ("Sara", 30)]
```

```python
Marks ≥ 85 → "Excellent"

Marks ≥ 60 → "Good"

Marks ≥ 40 → "Average"

Else → "Fail"
```

```python
for name, marks in students:
        if marks >= 85:
            print (name, "Excellent")
        elif marks >= 60:
            print (name,"Good")
        elif marks >= 40:
            print (name, "Average")
        else:
            print (name, "Fail")
    
```

```python
servers = {
    "Server1": 92,
    "Server2": 70,
    "Server3": 50,
    "Server4": 15
}
```

```python
for server, usage in servers.items():
    if usage >= 90:
        print (server, "Critical")
    elif usage >= 70:
        print (server, "High")
    elif usage >= 50:
        print (server, "Normal")
    else:
        print (server, "Idle")
```

```python
cart = [("Laptop", 85000), ("Keyboard", 2500), ("Mouse", 1200), ("Monitor", 15000)]
#If price ≥ 50000 → "10% Discount"

#If price between 10000–49999 → "5% Discount"

#If price < 10000 → "No Discount"
```

```python
for product, price in cart:
    match price:
        case p if p >= 50000:
            print ("10% Discount and the Final Price of",product, (10/100)*price)
        case p if p <= 49999 and p >= 10000:
            print ("5% Discount and the Final Price of ", product, (5/100)*price)
        case p if p < 10000:
            print ("No Discount" ,product, price)
            
```
