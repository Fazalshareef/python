```python
Instance

An instance is an object that belongs to a class. For instance, a list is a class in Python. 
    When we create a list, we have an instance of the list class.

 Instance variable: In class-based, object-oriented programming, an instance variable is a variable defined in a class (ie.. a member varaible),
   for which each instantiated object of the class has a separate copy, or instance.
```
