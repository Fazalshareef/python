```python
Dictionaries are unordered mappings for storing objects. Previously, we saw how lists store objects in an ordered sequence, and dictionaries use 
key-value pairing instead.
    This key value pair allows users to quickly grab objects without needing to know an index location.

    Dictionaries use curly braces and colons to signify the keys and their associated values.

    {'key1':'value1', 'key2':'value2'}

    Dictionaries are mutable
```

```python
Dictionary vs List

Dictionary is used when we don;t know the index values and where the items cannot be sorted. We just need to know the key value pair.

    Lists can be retrived based on location. Ordered sequence can be indexed and sliced

KEYS SHOULD ALWAYS BE STRINGS. VALUES CAN BE ANY DATA TYPE.
```
